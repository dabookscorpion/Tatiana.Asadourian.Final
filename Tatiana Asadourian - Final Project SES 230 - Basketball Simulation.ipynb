```python
from random import random
def basketball(teamA, teamB):
    intro(teamA, teamB)
    
    probA, probB, n = inputs()
    regwinsA, regwinsB, Ashutouts, Bshutouts = simNgames(n, probA, probB)
    summary(n,regwinsA,regwinsB,Ashutouts,Bshutouts)
    
def intro(team1, team2):
    print("This program will simulate a game of basketball between")
    print(team1, "and" team2 ". The abilities of each team are")
    print("indicated by their online statistics  that")
    print("that player will win the point when serving. Player A always")
    print("has the first serve.\n") 
    
def inputs():
    a = float(input("What is the probability that player A wins a serve? "))
    b = float(input("What is the probability that player B wins a serve? "))
    n = int(input("How many games would you like to simulate? "))
    return a, b, n

def simNgames(n, probA, probB):
    regwinsA = regwinsB = 0
    Ashutouts = Bshutouts = 0
    for i in range(n):
        scoreA, scoreB = simOneGame(probA, probB)
        if scoreA == 15:
            regwinsA = regwinsA + 1
        elif scoreA == 7 and scoreB == 0:
            Ashutouts = Ashutouts + 1
        elif scoreB == 15:
            regwinsB = regwinsB + 1
        elif scoreA == 0 and scoreB == 7:
            Bshutouts = Bshutouts + 1
    return regwinsA, regwinsB, Ashutouts, Bshutouts

def simOneGame(probA, probB):
    serving = "A"
    scoreA = 0
    scoreB = 0
    while not gameOver(scoreA, scoreB) and not shutOut(scoreA,scoreB):
        if serving == "A":
            if random() < probA:
                scoreA = scoreA + 1
            else:
                serving = "B"
        else:
            if random() < probB:
                scoreB = scoreB + 1
            else:
                serving = "A"
    return scoreA, scoreB

def gameOver(sA,sB):
    return sA == 15 or sB == 15

def shutOut(shA, shB):
    return (shA == 7 and shB == 0) or (shA == 0 and shB == 7)

def summary(n,rA,rB,sA,sB):
    print("Games simulated:",n)
    wA = rA + sA
    wB = rB + sB
    percentAwins = (wA/n)*100
    percentBwins = (wB/n)*100
    percentAshuts = (sA/wA)*100
    percentBshuts = (sB/wB)*100
    print("Wins for A: ",wA," (",percentAwins,"%)",sep="")
    print("Shutouts for A: ",sA," (",percentAshuts,"%)",sep="")
    print("Wins for B: ",wB," (",percentBwins,"%)",sep="")
    print("Shutouts for B: ",sB," (",percentBshuts,"%)",sep="")
```
