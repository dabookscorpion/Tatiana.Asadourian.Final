So, instead, I will run each quarter according to the average amount of possessions each team typically obtains each quarter, and use that as the limit to how long the quarter lasts. This will also account for many of the time-stopping aspects of the game like fouls and turnovers, which will be simulated in my program, so this alteration should still be relatively accurate to real games.

In real life, it's known that half-court shots are only attempted 25% of the time in the first, second, and third quarter, and less than 10% of the time in the fourth quarter. So, I will investigate not only how often the half-court shot succeeds (according to Wikipedia, the chance of making a half court shot is 1 in 100 attempts, so I will use this statistic) but in particular, I want to investigate how often this actually changes the outcome of a game that is very close in points at the end. In my program, I will run an attempt at initiating a half-court shot in the first, second, and third quarter with a 25% probability of a player attempting it and a 1% chance of making the shot. In the fourth quarter, I will only run the attempt if the points, after being assessed, only have a 1 to 3 point difference, warranting the urgency for a half-court shot in the last possession of the game.
